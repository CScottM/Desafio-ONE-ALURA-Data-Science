### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
# @title Frete

from matplotlib import pyplot as plt
loja['Frete'].plot(kind='hist', bins=20, title='Frete')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

#1. Análise do faturamento


```python

faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento total da Loja 1: {faturamento_loja1:.2f}")
print(f"Faturamento total da Loja 2: {faturamento_loja2:.2f}")
print(f"Faturamento total da Loja 3: {faturamento_loja3:.2f}")
print(f"Faturamento total da Loja 4: {faturamento_loja4:.2f}")
```

# 2. Vendas por Categoria


```python
# prompt: calcular a quantidade de produtos vendidos por categoria em cada loja.

import pandas as pd
vendas_categoria_loja1 = loja.groupby('Categoria do Produto')['Produto'].count().reset_index()
vendas_categoria_loja1.rename(columns={'Produto': 'Quantidade Vendida'}, inplace=True)
vendas_categoria_loja1['Loja'] = 'Loja 1'

vendas_categoria_loja2 = loja2.groupby('Categoria do Produto')['Produto'].count().reset_index()
vendas_categoria_loja2.rename(columns={'Produto': 'Quantidade Vendida'}, inplace=True)
vendas_categoria_loja2['Loja'] = 'Loja 2'

vendas_categoria_loja3 = loja3.groupby('Categoria do Produto')['Produto'].count().reset_index()
vendas_categoria_loja3.rename(columns={'Produto': 'Quantidade Vendida'}, inplace=True)
vendas_categoria_loja3['Loja'] = 'Loja 3'

vendas_categoria_loja4 = loja4.groupby('Categoria do Produto')['Produto'].count().reset_index()
vendas_categoria_loja4.rename(columns={'Produto': 'Quantidade Vendida'}, inplace=True)
vendas_categoria_loja4['Loja'] = 'Loja 4'

vendas_por_loja_categoria = pd.concat([vendas_categoria_loja1, vendas_categoria_loja2, vendas_categoria_loja3, vendas_categoria_loja4])

print("Quantidade de produtos vendidos por categoria em cada loja:")
vendas_por_loja_categoria
```

# 3. Média de Avaliação das Lojas

```python

media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de avaliação da Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de avaliação da Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de avaliação da Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de avaliação da Loja 4: {media_avaliacao_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
# prompt:  identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja.  Ao final, visualize os resultados de forma que fique claro quais produtos se destacaram nas vendas de cada loja.

lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

for nome_loja, df_loja in lojas.items():
  vendas_produto = df_loja.groupby('Produto')['Preço'].count().reset_index()
  vendas_produto.rename(columns={'Preço': 'Quantidade Vendida'}, inplace=True)

  produto_mais_vendido = vendas_produto.loc[vendas_produto['Quantidade Vendida'].idxmax()]
  produto_menos_vendido = vendas_produto.loc[vendas_produto['Quantidade Vendida'].idxmin()]

  print(f"--- {nome_loja} ---")
  print(f"Produto mais vendido: {produto_mais_vendido['Produto']} ({produto_mais_vendido['Quantidade Vendida']} unidades)")
  print(f"Produto menos vendido: {produto_menos_vendido['Produto']} ({produto_menos_vendido['Quantidade Vendida']} unidades)")
  print("\n")
```

# 5. Frete Médio por Loja


```python
# prompt: calcular o custo médio de frete para cada loja. O objetivo é entender quanto, em média, está sendo gasto com frete para cada uma das lojas.

media_frete_loja1 = loja['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

print(f"Custo médio do frete da Loja 1: {media_frete_loja1:.2f}")
print(f"Custo médio do frete da Loja 2: {media_frete_loja2:.2f}")
print(f"Custo médio do frete da Loja 3: {media_frete_loja3:.2f}")
print(f"Custo médio do frete da Loja 4: {media_frete_loja4:.2f}")
```

## Visualização dos produtos mais e menos vendidos por loja



```python
import matplotlib.pyplot as plt
import seaborn as sns

lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

most_sold_products = []
least_sold_products = []

for nome_loja, df_loja in lojas.items():
  vendas_produto = df_loja.groupby('Produto')['Quantidade de parcelas'].count().reset_index() # Count product occurrences as quantity sold
  vendas_produto.rename(columns={'Quantidade de parcelas': 'Quantidade Vendida'}, inplace=True)

  produto_mais_vendido = vendas_produto.loc[vendas_produto['Quantidade Vendida'].idxmax()]
  produto_mais_vendido['Loja'] = nome_loja
  most_sold_products.append(produto_mais_vendido)

  produto_menos_vendido = vendas_produto.loc[vendas_produto['Quantidade Vendida'].idxmin()]
  produto_menos_vendido['Loja'] = nome_loja
  least_sold_products.append(produto_menos_vendido)

df_most_sold = pd.DataFrame(most_sold_products)
df_least_sold = pd.DataFrame(least_sold_products)

print("Most Sold Products per Store:")
display(df_most_sold)

print("\nLeast Sold Products per Store:")
display(df_least_sold)

# Graph 1: Most Sold Products
plt.figure(figsize=(10, 6))
sns.barplot(data=df_most_sold, x='Loja', y='Quantidade Vendida', hue='Produto', dodge=False)
plt.title('Produtos Mais Vendidos por Loja')
plt.xlabel('Loja')
plt.ylabel('Quantidade Vendida')
plt.legend(title='Produto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Graph 2: Least Sold Products
plt.figure(figsize=(10, 6))
sns.barplot(data=df_least_sold, x='Loja', y='Quantidade Vendida', hue='Produto', dodge=False)
plt.title('Produtos Menos Vendidos por Loja')
plt.xlabel('Loja')
plt.ylabel('Quantidade Vendida')
plt.legend(title='Produto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Graph 3: Combined Most and Least Sold Products
df_most_sold['Tipo'] = 'Mais Vendido'
df_least_sold['Tipo'] = 'Menos Vendido'
df_combined = pd.concat([df_most_sold, df_least_sold])

plt.figure(figsize=(12, 6))
sns.barplot(data=df_combined, x='Loja', y='Quantidade Vendida', hue='Produto', palette='viridis', dodge=True)
plt.title('Produtos Mais e Menos Vendidos por Loja')
plt.xlabel('Loja')
plt.ylabel('Quantidade Vendida')
plt.legend(title='Produto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Relatório de Análise das Lojas para Decisão de Venda

## Introdução

Este relatório tem como objetivo analisar o desempenho de quatro lojas (Loja 1, Loja 2, Loja 3 e Loja 4) com base em dados de faturamento, vendas por categoria, avaliação dos clientes, produtos mais e menos vendidos e custo médio de frete. O propósito da análise é fornecer ao Senhor João informações embasadas para auxiliá-lo na decisão de qual loja vender.

## Análise dos Dados e Visualizações

### 1. Faturamento Total das Lojas

A análise do faturamento total revelou os seguintes valores:

*   **Loja 1:** R$ 1.534.509,12
*   **Loja 2:** R$ 1.488.459,06
*   **Loja 3:** R$ 1.464.025,03
*   **Loja 4:** R$ 1.384.497,58

Os gráficos de barras e pizza gerados anteriormente ilustram claramente que a Loja 1 apresenta o maior faturamento total entre as quatro lojas, enquanto a Loja 4 possui o menor faturamento.

### 2. Vendas por Categoria de Produto

A análise das vendas por categoria em cada loja, visualizada através dos gráficos de barras agrupado e empilhado, mostrou a distribuição das vendas por tipo de produto em cada unidade. Observamos variações na quantidade vendida de cada categoria entre as lojas, o que pode indicar diferentes perfis de clientela ou estratégias de estoque e marketing. A Loja 3 se destaca com a maior quantidade de móveis vendidos, enquanto a Loja 1 lidera nas vendas de eletrônicos.

### 3. Média de Avaliação dos Clientes

As médias de avaliação dos clientes para cada loja foram as seguintes:

*   **Loja 1:** 3.98
*   **Loja 2:** 4.04
*   **Loja 3:** 4.05
*   **Loja 4:** 4.00

Os gráficos de barras e dispersão da média de avaliação indicam que as Lojas 2 e 3 possuem as maiores médias de avaliação, sugerindo uma maior satisfação dos clientes nessas unidades. A Loja 1 tem a menor média de avaliação.

### 4. Produtos Mais e Menos Vendidos

A identificação dos produtos mais e menos vendidos em cada loja, apresentada anteriormente, revelou produtos de destaque e com menor saída em cada unidade. Por exemplo, "Guarda roupas" foi o mais vendido na Loja 1, enquanto "Celular ABXY" foi o menos vendido. Essa análise é crucial para entender o que impulsiona as vendas e o que pode estar com baixo desempenho em cada local.

### 5. Custo Médio do Frete

O custo médio do frete para cada loja foi calculado como:

*   **Loja 1:** R$ 34.69
*   **Loja 2:** R$ 33.62
*   **Loja 3:** R$ 33.07
*   **Loja 4:** R$ 31.28

Os gráficos de barras e dispersão do custo médio do frete demonstram que a Loja 1 possui o maior custo médio de frete, enquanto a Loja 4 apresenta o menor custo médio. Isso pode impactar a lucratividade de cada loja.

## Conclusão e Recomendação

Com base na análise abrangente dos dados, a **Loja 4** é a candidata mais forte a ser vendida.

**Justificativa:**

*   **Menor Faturamento:** A Loja 4 apresenta consistentemente o menor faturamento total entre as quatro lojas, o que a torna a menos rentável em termos de receita bruta.
*   **Custo Médio de Frete:** Embora tenha o menor custo médio de frete, o impacto positivo disso no faturamento geral é limitado devido ao baixo volume de vendas em comparação com as outras lojas.
*   **Produtos Menos Vendidos:** A análise dos produtos menos vendidos pode indicar problemas específicos de estoque, demanda ou concorrência na Loja 4 que contribuem para seu desempenho inferior.

As outras lojas (Loja 1, Loja 2 e Loja 3) apresentam faturamentos mais altos e, no caso das Lojas 2 e 3, médias de avaliação de clientes superiores. A Loja 1, apesar do maior faturamento, tem a menor média de avaliação e o maior custo médio de frete, o que pode ser um ponto de atenção, mas seu faturamento ainda a torna mais atrativa financeiramente do que a Loja 4.

Portanto, considerando o objetivo de vender uma loja e maximizar o retorno ou minimizar perdas, a Loja 4, com seu menor faturamento e desempenho geral menos expressivo, parece ser a opção mais lógica para a venda. Recomenda-se uma análise mais aprofundada dos motivos do baixo desempenho da Loja 4 para otimizar o processo de venda, se possível.
